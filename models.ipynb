```python
# imports 
import pandas as pd  # dataset
import seaborn as sns  #visualization (plottings)
import matplotlib.pyplot as plt  #visualization
import numpy as np  # arrays, algebra, dealing with numbers
import plotly.figure_factory as ff
import math 

from sklearn.model_selection import train_test_split # splitting test sets
from sklearn.metrics import classification_report, mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing
from sklearn.svm import OneClassSVM # anomoly
from numpy import where
```

```python
df = pd.read_csv('./Salary_Data.csv') # get data from cost.csv file
```

```python
# Data processing / Data cleaning
# , 'Annual Income ($)', 'Gender', 'Age', 'Work Experience', 'Family Size'
# Remove columns  
dfp = df.copy().drop(columns=[''])

dfp = pd.DataFrame(preprocessing.OrdinalEncoder().fit_transform(dfp), columns=dfp.columns)
print(dfp.columns[dfp.isna().any()])
dfp = dfp.fillna(0)


print("Columns size:", len(dfp))

print(dfp.head(100))
```

```python
# split train and test into 70:30
train, test = train_test_split(dfp, test_size=0.3, random_state=1)
x_train, y_train = train.drop(columns=['Spending Score (1-100)']), train['Spending Score (1-100)']
x_test, y_test = test.drop(columns=['Spending Score (1-100)']), test['Spending Score (1-100)']

print(x_train)
print("***********************")
print(y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
# Polynomial Model

print(x_train)
degreeList = [1,2,3,4,5,6,7,8,9,14,20]

for deg in degreeList:
    poly = PolynomialFeatures(degree=deg)
    
    x_poly = poly.fit_transform(x_train)
    plr = LinearRegression()
    
    plr.fit(x_poly, y_train)
    
    predicted_train = plr.predict(poly.transform(x_train))
    predicted_test = plr.predict(poly.transform(x_test))
    
    print("\n*********************************")
    print(f'Polynomial regression with degree = {deg}')
#     print("Model weights: ")
#     print(plr.coef_)

    rmse = np.sqrt(mean_squared_error(predicted_train, y_train))
    r2 = r2_score(y_train, predicted_train)
    print('\nMean sqaured error: ', mean_squared_error(predicted_train, y_train))
    print('\nTraining RMSE: %8.15f' % rmse)
    print('Training R2: %8.15f' % r2)

    # Reporting for the test set:
    rmse_test = np.sqrt(mean_squared_error(predicted_test, y_test))
    r2_test = r2_score(y_test,predicted_test)
    print('\nTest RMSE: %8.15f' % rmse_test)
    print('Test R2: %8.15f' % r2_test)

#     plt.scatter(x_train, y_train, s=10)
#     # sort the values of x before line plot
#     sort_axis = operator.itemgetter(0)
#     sorted_zip = sorted(zip(x_train, plr.predict(x_poly)), key=sort_axis)
#     x_poly_train, y_train_pred = zip(*sorted_zip)
#     plt.plot(x_poly_train, y_train_pred, color='m')
#     plt.show()
```
